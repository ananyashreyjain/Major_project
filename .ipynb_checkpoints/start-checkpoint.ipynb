```python
import gym
import random
import numpy as np
from statistics import mean, median
from collections import Counter
```

```python
LR = 1e-3
env = gym.make('CartPole-v0')
env.reset()
params = dict(
    goal_steps = 500,
    score_requirement = 50,
    initial_games = 10000,
    training_data = [],
    hot_encoded_columns = 2,
    score = []
)
```

```python
def get_data(params):
    env.reset()
    for game in range(params["initial_games"]):
        score = 0
        env.reset()
        game_data = []
        prev_observation = None
        for step in range(params["goal_steps"]):
            action = env.action_space.sample()
            observation, reward, done, info =  env.step(action)
            score += reward
            hot_encoded_action=[0]*params['hot_encoded_columns']
            hot_encoded_action[action] = 1
            if prev_observation is not None:
                game_data.append([prev_observation.tolist(), hot_encoded_action])
            prev_observation = observation
            if done:
                break
        if score>=params['score_requirement']:
            params['training_data'].append(game_data)
            params['score'].append(score)
get_data(params)
```

```python
X=[]
Y=[]
for game in params['training_data']:
    for move in game:
        X.append(move[0])
        Y.append(move[1])      
```

```python
Y
```

```python
import tensorflow as tf

model = tf.keras.Sequential([tf.keras.layers.Dense(128,input_shape=(6,),activation='relu'),
                             tf.keras.layers.Dense(256, activation='relu'),
                             tf.keras.layers.Dense(128, activation='relu'),
                             tf.keras.layers.Dense(2, activation='softmax'),
                                
])

model.compile(loss='spare_categorical_crossentropy', optimizer='adam')
model.fit(X,Y, epochs=1)
```
